```python
tabularView = {}

positionOfShipB = shipBRecords[0]
positionOfShipA = shipARecords[0]
for time in timeIndex:
    if len(shipA[shipA.BaseDateTime == time]):
        positionOfShipA = (shipA[shipA.BaseDateTime == time].iloc[0]['LAT'], shipA[shipA.BaseDateTime == time].iloc[0]['LON'])
    if len(shipB[shipB.BaseDateTime == time]):
        positionOfShipB = (shipB[shipB.BaseDateTime == time].iloc[0]['LAT'], shipB[shipB.BaseDateTime == time].iloc[0]['LON'])
    tabularView[time] = {"time": time, "shipA": positionOfShipA, "shipB": positionOfShipB}
```

```python
tabularView = []
shipStatusA= {
    "LAT": shipARecords[0][0],
    "LON": shipARecords[0][1],
    "COG": shipA.iloc[0].COG,
    "SOG": shipA.iloc[0].SOG,
    "Heading": shipA.iloc[0].Heading
}
shipStatusB = {
    "LAT": shipBRecords[0][0],
    "LON": shipBRecords[0][1],
    "COG": shipB.iloc[0].COG,
    "SOG": shipB.iloc[0].SOG,
    "Heading": shipB.iloc[0].Heading
}
for time in timeIndex:
    if len(shipA[shipA.BaseDateTime == time]):
        recordA = shipA[shipA.BaseDateTime == time].iloc[0]
        shipStatusA = recordA[["LAT","LON","COG","SOG","Heading", "MMSI"]]
    if len(shipB[shipB.BaseDateTime == time]):
        recordB = shipB[shipB.BaseDateTime == time].iloc[0]
        shipStatusB = recordB[["LAT","LON","COG","SOG","Heading", "MMSI"]]

    resultA = shipStatusA
    resultA['ship'] = "a"
    resultA['time'] = str(time)
    tabularView.append(resultA)
    resultB = shipStatusB
    resultB['ship'] = "b"
    resultB['time'] = str(time)
    tabularView.append(resultB)
```

```python
tabularView = pd.DataFrame(tabularView)
```

```python
tabShipA = tabularView[tabularView.ship == 'a']
tabShipB = tabularView[tabularView.ship == 'b']
```

```python
tabShipA.columns = "A_" + tabShipA.columns
tabShipA = tabShipA.rename(columns={"A_time": "time"})
tabShipA['time'] = pd.to_datetime(tabShipA['time'])
tabShipB.columns = "B_" + tabShipB.columns
tabShipB = tabShipB.rename(columns={"B_time": "time"})
tabShipB['time'] = pd.to_datetime(tabShipB['time'])
```

```python
tabShipA.join(tabShipB, on='time', how='right')
```
