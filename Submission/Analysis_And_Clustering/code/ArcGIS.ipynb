```python
raw = pd.read_csv('../data/AIS_ASCII_by_UTM_Month/2017/AIS_2017_01_Zone01.csv')
```

```python
single_ship_id = 352844000
```

```python
single_ship_records = raw[raw.MMSI == single_ship_id].sort_values(by="BaseDateTime")
```

```python
gis = GIS()
map1 = gis.map()
for r in single_ship_records.iterrows():
    record = r[1] 
    map1.draw([record.LAT, record.LON], {"title":record.BaseDateTime, "content":record.Status})
#     break
map1
```
